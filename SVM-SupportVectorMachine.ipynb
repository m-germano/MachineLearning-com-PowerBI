## **Código**

```python
import pickle

with open('bank.pkl', mode='rb') as f:
  x_train, y_train, x_test, y_test = pickle.load(f)
```

```python
from sklearn.svm import SVC

#Kernel = Radial Base Function -> Usado para separar os dados. RBF vem como default
#Random_state -> Fixa os dados para não ter diferença
#C -> penalização de falhas -> Se a penalização for muito grande, corre risco do modelo tentar se ajustar demais ao treinamento, o que gera o overfitting.
svm = SVC(kernel='rbf', random_state=1, C=2)
svm.fit(x_train, y_train)
previsoes = svm.predict(x_test)
previsoes
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(y_test, previsoes)
```

```python
print(classification_report(y_test, previsoes))
```
