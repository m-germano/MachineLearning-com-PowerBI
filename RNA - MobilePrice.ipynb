```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix

df_mobile_price = pd.read_csv('C:/Users/Matheus Germano/Downloads/mobile_price.csv')


#Dividindo colunas de dados da coluna de predição
x_mobile_price = df_mobile_price.drop('price_range', axis=1)
y_mobile_price = df_mobile_price['price_range']

#Guardar o dataframe original para reuso (antes do scaler)
x_mobile_price_original = x_mobile_price.copy()

#Normalizando escala das colunas numericos

scaler_mobile_price = StandardScaler()
x_mobile_price = scaler_mobile_price.fit_transform(x_mobile_price)

#Separar a base de treino da base de teste

x_mobile_price_treinamento,  x_mobile_price_producao, \
y_mobile_price_treinamento, y_mobile_price_producao, \
x_mobile_price_original_treinamento, x_mobile_price_original_producao \
= train_test_split(x_mobile_price, y_mobile_price, x_mobile_price_original, test_size=0.2, random_state=0)

#Rodar a Rede neural

mlp_mobile_price = MLPClassifier(max_iter=1000, tol=0.000010, solver='adam', hidden_layer_sizes=(5,5,5,5), activation='relu',random_state=0)
mlp_mobile_price.fit(x_mobile_price_treinamento, y_mobile_price_treinamento)
previsoes = mlp_mobile_price.predict(x_mobile_price_producao)

accuracy_score(y_mobile_price_producao, previsoes)
print(classification_report(y_mobile_price_producao, previsoes))

#Criando dataframe de resultados
df_resultado = x_mobile_price_original_producao.copy()
df_resultado['price_range_real'] = y_mobile_price_producao
df_resultado['price_range_predito'] = previsoes
df_resultado.head()

#Criando um dataframe de métricas

df_classification_report = pd.DataFrame(classification_report(y_mobile_price_producao, previsoes, output_dict=True)).transpose().reset_index().rename(columns={'index':'Classe'}).round(2)
df_classification_report


#Criando Dataframe de matriz de confusão
df_matriz_confusao = pd.DataFrame(confusion_matrix(y_mobile_price_producao, previsoes), columns=['Predito_0', 'Predito_1','Predito_2','Predito_3'], index=['Real_0','Real_1','Real_2','Real_3',]).reset_index()\
.rename(columns={'index':'Classe Real'})
df_matriz_confusao
```
