# Redes Neurais Artificiais - MKT Bank

```python
import pickle

with open('/content/drive/MyDrive/Machine Learning/EstudoModelos/BaseDeDados/bank.pkl', mode='rb') as f:
  x_train,y_train,x_test,y_test = pickle.load(f)
```

```python
from sklearn.neural_network import MLPClassifier

#solver Ã© um otimizador que auxilia nos pesos
mlp_bank = MLPClassifier(verbose=False, max_iter=1000, tol=0.0000010 ,solver='adam', hidden_layer_sizes=(4,4,4), activation='relu')

mlp_bank.fit(x_train, y_train)
previsoes = mlp_bank.predict(x_test)
previsoes
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(y_test, previsoes)
```

```python
print(classification_report(y_test, previsoes))
```
